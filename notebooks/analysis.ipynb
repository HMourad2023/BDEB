#### Valeurs catégorielles:

**Redondances des colonnes:**
- 'DEST_COUNTRY_NAME' / 'DEST_COUNTRY'
- 'ORIGIN_CITY_NAME' / 'ORIGIN_COUNTRY'
- 'ORIGIN_CITY_NAME' / 'ORIGIN'
- 'DEST' / 'DEST_CITY_NAME'
- 'UNIQUE_CARRIER' / 'UNIQUE_CARRIER_NAME' / 'CARRIER' / 'CARRIER_NAME'

```python
def plot_categories(cat_data, columns, n=10, palette='muted', row=5, col=4, figsize=(24,30)):
    fig, axes = plt.subplots(row, col, figsize=figsize)
    axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]
    
    for i, column in enumerate(columns):
        if i >= len(axes):
            break
        
        # Get top n unique categories by frequency
        order = cat_data[column].value_counts().index[:n]
        
        # Plot count plot
        sns.countplot(data=cat_data, y=column, order=order, palette=palette, ax=axes[i])
        
        # axes[i].set_title(f'Distribution des {column} (TOP-{n})')
        # axes[i].set_xlabel('Nombre de Vols')
        # axes[i].set_ylabel(column)
    
    # Hide any remaining unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    plt.savefig("../visualizations/distributions_des_variables_categorielles.jpg")
    plt.show()
```

```python
columns = cat_data.columns.tolist()
plot_categories(cat_data, columns)
```
